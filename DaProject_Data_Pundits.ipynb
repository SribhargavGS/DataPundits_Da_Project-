```python
fig, ((ax1),(ax2),(ax3))=plt.subplots(ncols=1, nrows=3)
fig.set_size_inches(30,20)
fig.tight_layout(pad=4.0)

pltplot(mean_for_yearly,'started','subscribers','lightcoral', ax1,'Subscribers per Year (mean)')
pltplot(mean_for_yearly,'started','video views','green', ax2,'Video views per Year (mean)')
pltplot(mean_for_yearly,'started','video count','gold', ax3,'Video count per Year (mean)')
```
